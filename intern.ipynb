```python
df['Aggregate rating'] = pd.to_numeric(df['Aggregate rating'], errors='coerce')
```

```python
print(df['Aggregate rating'].dtype)
```

```python
print(df['Aggregate rating'].describe())
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Aggregate rating'], kde=True, bins=30)
plt.title('Distribution of Aggregate Rating')
plt.xlabel('Aggregate Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['Aggregate rating'])
plt.title('Boxplot of Aggregate Rating')
plt.xlabel('Aggregate Rating')
plt.show()
```

```python
bins = [0, 2.5, 3.5, 4.5, 5]
labels = ['Poor', 'Average', 'Good', 'Excellent']
df['Rating Category'] = pd.cut(df['Aggregate rating'], bins=bins, labels=labels, include_lowest=True)
```

```python
rating_distribution = df['Rating Category'].value_counts(normalize=True) * 100
print(rating_distribution)
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Rating Category', data=df, order=labels)
plt.title('Distribution of Rating Categories')
plt.xlabel('Rating Category')
plt.ylabel('Count')
plt.show()
```

```python
from imblearn.over_sampling import SMOTE
df_filled['Rating Category'] = df['Rating Category']
X = df_filled.drop(columns=['Aggregate rating', 'Rating Category'])
y = df_filled['Rating Category']
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)
print(pd.Series(y_resampled).value_counts(normalize=True) * 100)
```
