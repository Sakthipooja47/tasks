```python
import numpy as np
import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/Dataset .csv")
df.head()
```

```python
numerical_summary=df.describe()
```

```python
numerical_df = df.select_dtypes(include=np.number)
```

```python
numerical_summary.loc['mean']=numerical_df.mean()
```

```python
numerical_summary.loc['median']=numerical_df.median()
```

```python
numerical_summary.loc['std']=numerical_df.std()
```

```python
print("Numerical Summary:")
print(numerical_summary)
```

```python
categorical_columns=['Country Code','City','Cuisines']
for col in categorical_columns:
  print(f"\nDistribution of {col}:")
  print(df[col].value_counts())
```

```python
top_cuisines=df['Cuisines'].value_counts().head(10)
top_cities=df['City'].value_counts().head(10)
```

```python
print("\nTop 10 Cuisines:")
print(top_cuisines)
```

```python
print("\nTop 10 Cities with Highest Number of Restaurents:")
print(top_cities)
```
